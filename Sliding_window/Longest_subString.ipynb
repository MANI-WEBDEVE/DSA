```python
s = 'muhammad'
print(len(s))
seen = {}
left = max_len = 0
for right, char in enumerate(s):
    print(f'right = {right} , char = {char}')
    if char in seen and seen[char] >= left:
        print(f'inter')
        left = seen[char] + 1
        print(f'left in loop  = {left}')
        print(f'max-len in loop = {max_len}')
    print('second inter')
    seen[char] = right
    max_len = max(max_len, right - left + 1)

print(seen)
```

```python
def longest_Substring(s):
  max_len_2 = 0

  for i in range(len(s)):
    print(f'i = {s[i]}')
    for j in range(i, len(s)):
      print(f'j = {s[j]}')
      if len(set(s[i:j+1])) == len(s[i:j+1]):
        max_len_2 = max(max_len_2, len(s[i:j+1]))

  return max_len_2


longest_Substring('muhammad')
```
